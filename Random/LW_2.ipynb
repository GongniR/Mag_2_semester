```python
def recup():
    save_recup = []
    for t in range(1, 1000):
        Sn = 0
        iter = 0
        while Sn<=t:
            csi =  abs(random.randint(0,10))
            Sn+=csi
            print(Sn)
            iter += 1
            if Sn>t:
                save_recup.append(iter-1)

    return save_recup
    
```
