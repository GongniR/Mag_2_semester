```python
def strax_Spare_Anderson(y0=10,c=1):
    save_recup = []
    Yt_list = []
    
    for t in range(0, 100):
        Yt = y0
        Sn = 0
        iter = 0
        while Sn<=t:
            csi =  abs(random.randint(0,4))
            Sn+=csi

            iter += 1
            if Sn>t:
                save_recup.append(iter-1)
        x_r = [random.randint(-10,10) for i in range(save_recup[t])]
        # print(sum(x_r))
        Yt += c*t+sum(x_r)
        
        Yt_list.append(Yt)
    return Yt_list   
```

```python
Yt_1 = strax_Spare_Anderson()
x_1 = [i for i in range(len(Yt_1))]
print(Yt_1)
plt.plot(x_1,Yt_1)
```

```python
Yt_2 = strax_Spare_Anderson()
x_2 = [i for i in range(len(Yt_2))]
print(Yt_2)
plt.plot(x_2,Yt_2)
```

```python
mu_1 = np.array(Yt_1).mean()
mu_2 = np.array(Yt_2).mean()
print("Мат. ожидание 1 выборки:", mu_1)
print("Мат. ожидание 2 выборки:", mu_2)
```
